```python
city = ["up","rj","mh","mp"]
population = [15000,11243,9008,4500]
colours = ["green",'black','pink','skyblue']

##bar plot

plt.bar(city,population,color=colours,edgecolor = "black",width=0.2,label=city)
# plt.show()
plt.legend()
```

```python
###histogram 
##intervals - 0-7  -- bins 

age = np.random.randint(5,90,250)

plt.hist(age,bins = 25)

plt.show()

```

```python
##pieplot

labels = ['Apple','Banana','Cherry']
sizes = [20,30,50]

plt.figure(figsize=(8,8))

plt.pie(sizes, labels = labels, autopct = "%1.1f%%")
plt.savefig("pieplot.png")
plt.show()
```
