```python
##outliers -- execption(kuch points jo diff behave karte hai from most of the data points)
##aapna ko outliers se deal karna padhta hai --- remove and scale aand transform 
```

```python
upper_limit = df["total_bill"].quantile(0.70)
sns.boxplot(df[df["total_bill"]<upper_limit]["total_bill"])
##quantile means intervall 
```

```python
##scale and tranform 
q1 = np.percentile(df["total_bill"],25)
q2 = np.percentile(df["total_bill"],75)
q1,q2
```

```python
IQR = q2-q1
IQR
```

```python
upperr_limit = q2 + 1.5 *IQR
lowerr_limit = q1 -1.5*IQR

upperr_limit,lowerr_limit
```

```python
# outliers_free_bill = df[(df["total_bill"]<=upperr_limit) & df(df["total_bill"]>=lowerr_limit)]

# outliers_free_bill
```

```python
# df[(df["total_bill"]>=upperr_limit) & df(df["total_bill"]>=lowerr_limit)]
```

```python
outliers_free_bill= df[(df["total_bill"] >= lowerr_limit) & (df["total_bill"] <= upperr_limit)]

outliers_free_bill
```

```python
sns.boxplot(outliers_free_bill['total_bill'])
```
